```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')
data_path = 'datasets/data.yaml'
model.train(data=data_path,epochs=50, imgsz=640)
```

```python
result = model.predict('mnist/000800.jpg')
result[0].show()
```

```python
data_path = 'datasets/data.yaml'
metrics1 = model.val(data=data_path)
```

```python
metrics1
```

```python
model = YOLO('yolov8n.pt')
data_path = 'datasets/data.yaml'
model.tune(data=data_path, epochs=50, imgsz=640, iterations=10, val=True)
```

```python
model = YOLO('runs/detect/tune6/weights/best.pt')
result = model.predict('mnist/000999.jpg')
result[0].show()
```
